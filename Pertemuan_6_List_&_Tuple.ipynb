## List / Array

List adalah tipe data yang berurut (sequensial) yang mana didalamnya berisi nilai satu atau lebih elemen.

Tanda list adalah menggunakan kurung `[ ]`.

```python
# my_list = []
# my_list merupakan nama variabel
# [] merupakan nilai dari list
#  didalam [] inilah yang disebut elemen/item

my_list_angka = [
                 200, #indeks ke 0
                 400, #indeks ke 1
                 600 #indeks ke 2
                 ] #numerik/integer
# my_list_huruf = ["Hallo", "Hai","UNM"]
# my_list_campuran = [1, "Hallo", 3.14]

# my_list_python = ['P','Y','T','H','O','N','I','N','D','O']
print(my_list_angka)
# indeks = angka yang selalu dimulai dari nol (0)
# print(my_list_python[4])
# Kita bisa pakai range / jangkauan untuk menampilkan nilai yang ada sesuai range (:) yang kita tentukan
# print(my_list_python[:6])
```

### Menambahkan Anggota pada List

**fungsi append()**

```python
my_list_data1 = ["Nasi Goreng", "Mie Goreng"]

my_list_data1.append("Kwetiau")

print(my_list_data1[2])
```

**fungsi insert()**

```python
my_list_data2 = [1,2,3,4,6,7,8,9]

# insert mesti hitung indeks juga
my_list_data2.insert(4, 5)
print(my_list_data2)
```

**fungsi remove()** -> Hapus data dengan memasukkan nilai yang mau dihapus

```python
my_list_remove = ['a','b','c','d','e','f','g',3.14]
my_list_remove.remove(3.14)
print(my_list_remove)
```

**fungsi pop()** -> hapus data dengan memasukkan nilai indeks yang mau dihapus

```python
my_list_pop = ['a','b','c','d','e','f','g',3.14]
my_list_pop.pop(0)
my_list_pop.pop(1)
my_list_pop.pop(2)
my_list_pop.pop(3)
# my_list_pop.pop(4)
print(my_list_pop)
```

**fungsi clear()**

```python
my_list_clear = ['a','b','c','d','e','f','g',3.14]
my_list_clear.clear()
print(my_list_clear)
```

**fungsi sort()**

```python
my_list_sort = ['a','b','c','d','e','f','g','z','x','v','m']
# sort selalu mulai dari yang terkecil sampai yang terbesar
my_list_sort.sort()
print(my_list_sort)
```

```python
my_list_sort.sort(reverse=True)
print(my_list_sort)
```

## Tuple

**Tuple** bersifat Immutable yang mana artinya data/nilai yang sudah dideklarasikan tidak bisa diganggu gugat.

**List** bersifat mutable yang mana artinya data/nilai yang sudah dideklarasikan dapat kita modifikasi (append, insert, remove, dll)

```python
my_tuple = ()
# my_tuple merupakan nama variabel
# () merupakan nilai dari tuple
#  didalam () inilah yang disebut elemen/item

my_tuple_angka = (
                 200, #indeks ke 0 dari tuple
                 400, #indeks ke 1 dari tuple
                 600, #indeks ke 2 dari tuple
                  [
                   #tipe data list yang ada didalam tuple
                   800, #indeks ke 0 dari list
                   1000 #indeks ke 1 dari list
                   ] #indeks ke 3 dari tuple
 ) #numerik/integer
# my_tuple_angka[3][0] = 700
# my_tuple_angka[3][1] = 900
print(250 in my_tuple_angka)
```

**Perulangan untuk Tuple/List**

```python
nama = ["Galih", "Ratna"]
for name in nama:
  print('Hi',name)
```

**Fungsi count()**

```python
my_tuple_count = ('P','Y','T','H','O','N','I','N','D','O')
print(my_tuple_count)
```

```python
# pemanggilan nilai berlaku case sensitif (Huruf besar kecilnya berpengaruh)
my_tuple_count.count('O')
```

**Fungsi index()**

```python
my_tuple_count.index('O')
```

**Studi Kasus 2**

```python
#variable yg berulang menggunakan List/matriks
list_nim=[]
list_uts=[]
list_uas=[]
list_total=[]
ulang=2
for i in range(ulang):
print ("data Ke - " + str(i+1))
list_nim.append(input("Masukkan Nim anda : "))
list_uts.append(int(input("Masukkan Nilai UTS anda:")))
list_uas.append(int(input("Masukkan Nilai UAS : ")))

#proses
for i in range(ulang):
list_total.append((list_uas[i] + list_uts[i]) / 2)

#Cetak
print("=============================================================")
print("Nim Nilai Uts Nilai UAS Total")
print("=============================================================")
for i in range(ulang):
print ("%s \t %i \t\t %i \t\t\t %i" % (list_nim[i],list_uts[i],list_uas[i],list_total[i]))
print("=============================================================")
```
