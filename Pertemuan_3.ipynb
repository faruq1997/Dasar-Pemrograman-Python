# Operator Logika dalam Pemrograman Python

**Operator** didalam Python adalah simbol khusus yang berfungsi untuk menjalankan suatu operasi tertentu, baik operasi aritmatika maupun operasi logika. Sedangkan nilai yang dioperasikan oleh operator dinamakan sebagai operand

Macam-Macam operator:

## 1. Operator Aritmatika

Jenis Operator Aritmatika:

operator aritmatika.PNG

```python
# Operator aritmatika biasanya yang umum digunakan, contoh program dengan aritmatika, Kalkulator

a = 10
b = 3

print(a, '+', b, '=', a + b)
print(a, '-', b, '=', a - b)
print(a, '*', b, '=', a * b)
print(a, '/', b, '=', a / b)
print(a, '%', b, '=', a % b)
print(a, '**', b, '=', a ** b)
print(a, '//', b, '=', a //b)
```

## 2. Operator Komparasi (Perbandingan)

Operator perbandingan adalah operator yang bertugas untuk membandingkan antar dua operan. Apakah operand 1 lebih besar dari pada operand 2, atau apakah keduanya sama? Dan lain sebagainya. Berikut simbol-simbol operator perbandingan

operator perbandingan.PNG

```python
a = 5
b = 10

print("Apakah...")
print(a, '>', b, '=', a > b)
print(a, '<', b, '=', a < b)
print(a, '==', b, '=', a == b)
print(a, '!=', b, '=', a != b)
print(a, '>=', b, '=', a >= b)
print(a, '<=', b, '=', a <= b)
```

## 3. Operator Penugasan

Operator penugasan adalah operator yang digunakan untuk memberikan sebuah tugas terhadap suatu variabel. Atau dalam bahasa yang lebih manusiawi: **operator penugasan adalah operator yang berfungsi untuk memberikan nilai ke dalam sebuah variabel.**

Sebenarnya operator penugasan ini hanya ada 1 saja, yaitu operator = .
Berikut simbol untuk operator ini:

operator penugasan.PNG


```python
# penugasan pertama
a = 10
# variabel a ini nantinya akan terus dijumlahkan sesuai nilai akhirnya
print('a = 10 -> ', a)

a += 5
print('a += 5 -> ', a)

a -= 3
print('a -= 3 -> ', a)

a *= 6
print('a *= 6 -> ', a)

a /= 8
print('a /= 8 -> ', a)

# karena a jadi float, kita ubah lagi menjadi integer
a = int(a)

a %= 9
print('a %= 9 -> ', a)

a //= 6
print('a //= 6 -> ', a)

a **= 1
print('a **= 1 -> ', a)


a &= 2
print('a &= 2 -> ', a)

a |= 3
print('a |= 3 -> ', a)

a ^= 4
print('a ^= 4 -> ', a)

a >>= 4
print('a >>= 4 -> ', a)

a <<= 2
print('a <<= 4 -> ', a)

# catatan: beberapa operator terakhir di atas mengandung operasi bitwise. Operasi bitwise ada di bahasan selanjutnya.
```

## 4. Operator Logika

Operator logika adalah operator yang sangat penting. Operator ini sangat berkaitan erat dengan operator perbandingan. Dan kedua-duanya juga mengembalikan nilai dengan tipe data yang sama yaitu *boolean*.
Berikut ini tabel dari operator logika pada python.

operator logika.PNG

```python
print(True and True)
print(1 + 2 == 3 and True)
print('----')
print(False or 1 > 5)
print(False or 5 > 2)
print('----')
print(not(1 > 5))
print(not(1 < 5))
```

## 5. Operator Bitwise

Operator bitwise adalah operator yang berhubungan dengan angka-angka biner.

Angka-angka biner adalah angka 0 dan 1. Dan pada hakikatnya hanya ini lah angka yang dipahami oleh mesin.

Sebelum kita mulai, kita bisa mengetahui nilai biner dari suatu angka desimal dengan melakukan perintah **format()** dengan parameter kedua berupa string **'08b'**.
(Sebagai komparasi, bisa juga bandingkan dengan tools ini: [Konverter Desimal ke Biner](https://www.rapidtables.org/id/convert/number/decimal-to-binary.html)

Misal:

```python
# biner dari angka 0
print(format(0, '08b'))
# biner dari angka 1
print(format(1, '08b'))
# biner dari angka 2
print(format(2, '08b'))
# biner dari angka 37
print(format(37, '08b'))
```

Operator bitwise ini operator yang melakukan operasi bit terhadap operand. Operator ini beroperasi bit per bit sesuai dengan namanya.
Sebagai misal, angka 2 dalam bit ditulis 10 dalam notasi biner dan angka 7 ditulis 111. Pada tabel di bawah ini, misalkan a = 10 (00001010) dalam biner dan b = 4 (00000100) dalam biner.
Setelah sedikit perkenalan dengan binary, berikut ini adalah tabel yang menjelaskan tentang operator bitwise pada python.

**Jenis Operator Bitwise**

operator bitwise.PNG

```python
a = 1
b = 64

print('a =', a, '=', format(a, '08b'))
print('b =', b, '=', format(b, '08b'), '\n')

print('[and]')
print('a & b =', a & b)
print(format(a, '08b'), '&', format(b, '08b'), '=', format(a & b, '08b'), '\n')

print('[or]')
print('a | b =', a | b)
print(format(a, '08b'), '|', format(b, '08b'), '=', format(a | b, '08b'), '\n')

print('[xor]')
print('a ^ b =', a ^ b)
print(format(a, '08b'), '^', format(b, '08b'), '=', format(a ^ b, '08b'), '\n')

print('[not]')
print('~a ~b =', ~a, ~b)
print('~' + format(a, '08b'), '~' + format(b, '08b'), '=', format(~a, '08b'), format(~b, '08b'), '\n')

print('[shift right]')
print('a >> b =', a >> b)
print(format(a, '08b'), '>>', format(b, '08b'), '=', format(a >> b, '08b'), '\n')

print('[shift left]')
print('b << a =', b << a)
print(format(b, '08b'), '<<', format(a, '08b'), '=', format(b << a, '08b'), '\n')
```

##6. Operator Keanggotaan

Python adalah bahasa pemrograman yang terbilang unik, ia memiliki operator khusus atau juga sering dinamakan sebagai operator spesial. Dinamakan spesial karena memang operator ini hanya ada di Python dan tidak ada di bahasa pemrograman lainnya.

Di antara operator spesial tersebut adalah operator keanggotaan. Atau didalam bahasa inggris ia dinamakan membership operator.

Operator keanggotaan dalam python hanya memiliki dua varian, yaitu **in** dan **not in**. Berikut ini tabel lebih lengkapnya:

operator keanggotaan.PNG

```python
kata = 'Indonesia'
print('Kata dasar :',kata)
print('\'p\' apakah in(ada) di '+kata+':', 'p' in kata)
print('\'a\' apakah in(ada) di '+kata+':', 'a' in kata)
print('\'p\' apakah not in(tidak ada) di '+kata+':', 'p' not in kata)
print('\'a\' apakah not in(tidak ada) di '+kata+':', 'a' not in kata)
print('\n')

# Bila tipe datanya array
abc = ['a','b','c']
print('abc :',abc)
print('\'a\' in abc     :', 'a' in abc)
print('\'a\' not in abc :', 'a' not in abc)
print('\'d\' not in abc :', 'd' not in abc)
print('\n')

# Bila tipe datanya list
angka = (12,43,102,55)
print('angka :',angka)
print('102 in angka     :', 102 in angka)
print('102 not in angka :', 102 not in angka)
print('35 not in angka  :', 35 in angka)
```

## 7. Operator Identitas

Selain operator keanggotaan, python juga masih memiliki operator spesial lainnya: yaitu operator identitas. Operator ini didefinisikan dengan **is** dan **is not**.

Tugasnya adalah untuk mengetahui apakah dua buah variabel memiliki nilai yang sama dan posisi yang sama di dalam memori. Karena tidak semua nilai yang sama memiliki tempat / posisi yang sama di dalam memori.

Jenis-jenisnya:

operator identitas.PNG

```python
# dalam python penulisan variabel bisa dipisah seperti ini:
# a = 5
# b = 5
# c = 6

# bisa juga digabung dengan posisi yang terurut dipisah dengan koma (,)
a,b,c = 5,5,6

print('a is b :', a is b)
print('a is c :', a is c)
print('a is not c :', a is not c)
print('\n')
  
i = 'Indonesia'
j = 'Malaysia'
print('i is j :', i is j)
print('i is not j :', i is not j)
print('\n');
  
x = ['a','b','c']
y = ['a','b','c']
print('x is y :', x is y)
print('x is not y :', x is not y)
```
