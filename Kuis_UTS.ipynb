```python
# awal inputan
print("PROGRAM BELI TOP-UP DIAMOND")
print("Toko UpinIpin Cell")
nama = input("Nama Pelanggan: ")
nope = input("Nomor Handphone: ")
jns_game = input("Jenis Game [ML/FF/AOV/PUBGM/POU]: ")
tpe_paket = input("Tipe Paket [A/B/C/D/E/F/G]: ")
jumlah_beli = int(input("Jumlah Beli: "))
no_voucher = int(input("Masukkan Kode Voucher untuk dapat diskon/diisi 0 jika kosong: "))
# akhir inputan

# awal Proses
a = 1500000
a_jml = 1500
b = 1200000
b_jml = 775
c = 1000000
c_jml = 625
d = 400000
d_jml = 242
e = 200000
e_jml = 121
f = 50000
f_jml = 30
g = 25000
g_jml = 15
kode_voucher = 14045

# Awal Jenis Game
if jns_game == "ML":
  nama_game = "Mobile Legends"
elif jns_game == "FF":  
  nama_game = "Free Fire"
elif jns_game == "AOV":  
  nama_game = "Arena of Valor"
elif jns_game == "PUBGM":  
  nama_game = "PUBG Mobile"
else:
  nama_game = "POU"
# Akhir Jenis Game

# Tipe Paket
if tpe_paket == "A":
  nama_paket = "Presiden"
  total_beli = jumlah_beli * a
elif tpe_paket == "B":
  nama_paket = "Gubernur"
  total_beli = jumlah_beli * b
elif tpe_paket == "C":
  nama_paket = "Bupati"
  total_beli = jumlah_beli * c
elif tpe_paket == "D":
  nama_paket = "Camat"
  total_beli = jumlah_beli * d
elif tpe_paket == "E":
  nama_paket = "Lurah"
  total_beli = jumlah_beli * e
elif tpe_paket == "F":
  nama_paket = "RW"
  total_beli = jumlah_beli * f
elif tpe_paket == "G":
  nama_paket = "RT"
  total_beli = jumlah_beli * g
else:
  total_beli = 0
  nama_paket = "Tidak ada tipe paket yang kamu pilih"
# akhir tipe paket

# harga paket
harga_paket = total_beli / jumlah_beli

# Voucher
if no_voucher == kode_voucher:
  # total bayar
  diskon = total_beli * 0.3
else:
  diskon = 0

# Total Bayar - Diskon
total_bayar = total_beli - diskon

# pajak
ppn = 0.1 * total_bayar

# Grant Total
grant_total = total_bayar + ppn

# akhir proses

# awal output
print("\n")
print("=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=* \n")
print("Toko UpinIpin Cell")
print("Nama Pelanggan:",nama)
print("Nomor Handphone:",nope)
print("\n")
print("=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*")
print("Selamat Pembelian Kamu Sukses.")
print("Jenis Game yang telah dipilih: "+ nama_game + "["+jns_game+"]")
print("Tipe Paket: ", tpe_paket)
print("Nama Paket: ", nama_paket)
print("Harga Paket: ", harga_paket)
print("Jumlah Beli:", jumlah_beli)
print("Total Harga:", total_beli)
print("Diskon(30%):", diskon)
print("Total Harga - Diskon:", total_bayar)
print("PPN (10%):", ppn)
print("Total Bayar:")
print("Rp.", grant_total)
print("=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=* \n")
print("Terima Kasih telah belanja.")
# akhir output
```
